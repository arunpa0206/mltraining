```python
def error(b):
    return (b-4)**2

```

```python
def slope(b):
    return 2*(b-4)


```

```python

b=108
lr=0.01
target = 4
max_iterations = 10000

def train(b,lr,target,max_iterations):
    for i in range(max_iterations):
        current_error=error(b)
        if(current_error==0):
            break
        current_slope=slope(b)
        b=b-(current_slope*lr)
        #print(b)
    return b
        
b=train(b,lr,target,max_iterations) 
print (b)
```
